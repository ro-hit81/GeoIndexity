```python
import ee
```

```python
ee.Authenticate()
ee.Initialize()
```

```python
random_point = ee.Geometry.Point([84, 37.38])
aoi = random_point.buffer(30)
image_collection = ee.ImageCollection("LANDSAT/LC08/C02/T1_TOA").filterDate("2020-01-01", "2021-01-01").filterBounds(aoi)
```

```python
def NDVI(image):
    red = image.select('B4')
    nir = image.select('B5')
    
    ndvi = nir.subtract(red).divide(nir.add(red)).rename('NDVI')
    
    return image.addBands(ndvi)
```

```python
ndvi_collection = image_collection.map(NDVI)
```

```python
def mean_NDVI(image):
    date = ee.Date(image.get('system:time_start')).format("YYYY-MM-dd", 'UTC')
    mean = image.select('NDVI').reduceRegion(
        reducer=ee.Reducer.mean(),
        geometry=aoi
    ).get('NDVI')
    return ee.Feature(None, {
        'date' : date,
        'mean_ndvi': mean
    })
```

```python
mean_ndvi_features = ndvi_collection.map(mean_NDVI).getInfo()
mean_ndvi_features
```

```python
dates = [feature['properties']['date'] for feature in mean_ndvi_features['features']]
mean_ndvi = [features['properties']['mean_ndvi'] for features in mean_ndvi_features['features']]
```

```python
import pandas as pd

df = pd.DataFrame({
    'Date': dates,
    'Mean_NDVI': mean_ndvi
})
df
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.plot(df['Date'], df['Mean_NDVI'], marker='o', linestyle='--')
plt.title('Mean NDVI Time Series')
plt.xlabel('Date')
plt.xticks(rotation=45)
plt.ylabel('Mean NDVI')
plt.grid(True)
plt.show()
```
